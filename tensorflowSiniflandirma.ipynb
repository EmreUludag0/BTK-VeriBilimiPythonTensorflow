```python
y = dataFrame["Type"].values
x = dataFrame.drop("Type",axis= 1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=15) 
```

```python
from sklearn.preprocessing import  MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
x_train.shape
```

```python
model = Sequential()

model.add(Dense(units= 30, activation="relu"))
model.add(Dense(units= 15, activation="relu"))
model.add(Dense(units= 15, activation="relu"))
model.add(Dense(units= 1, activation="sigmoid")) #0 ile 1 arasi deger verir.

model.compile(loss = "binary_crossentropy", optimizer = "adam")
```

```python
model.fit(x = x_train, y=y_train, epochs=700 ,validation_data=(x_test,y_test), verbose=1) #bard
```

```python
model.history.history
```

```python
modelKaybi = pd.DataFrame(model.history.history)
```

```python
modelKaybi.plot()
```

```python
model = Sequential()

model.add(Dense(units= 30, activation="relu"))
model.add(Dense(units= 15, activation="relu"))
model.add(Dense(units= 15, activation="relu"))
model.add(Dense(units= 1, activation="sigmoid")) #0 ile 1 arasi deger verir.

model.compile(loss = "binary_crossentropy", optimizer = "adam")
```

```python
EarlyStopping = EarlyStopping(monitor="val_loss", mode="min",verbose=1, patience=25) #bard
```

```python
model.fit(x=x_train,y=y_train, epochs=700, validation_data=(x_test,y_test),verbose=1, callbacks=[EarlyStopping]) #bard
```

```python
modelKaybi = pd.DataFrame(model.history.history)
```

```python
modelKaybi.plot()
```

DROPOUT

```python
model = Sequential()

model.add(Dense(units= 30, activation="relu"))
model.add(Dropout(0.6))

model.add(Dense(units= 15, activation="relu"))
model.add(Dropout(0.6))

model.add(Dense(units= 15, activation="relu"))
model.add(Dropout(0.6))

model.add(Dense(units= 1, activation="sigmoid")) #0 ile 1 arasi deger verir.

model.compile(loss = "binary_crossentropy", optimizer = "adam")
```

```python
model.fit(x=x_train,y=y_train, epochs=700, validation_data=(x_test,y_test),verbose=1, callbacks=[EarlyStopping]) 
```

```python
kayipDf = pd.DataFrame(model.history.history)
```

```python
kayipDf.plot()
```
