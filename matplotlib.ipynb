```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
yasListesi = [10,20,30,40,50,60,70]
kiloListesi = [20,60,80,90,25,45,10]
```

```python
numpyYasListesi = np.array(yasListesi)
numpyKiloListesi = np.array(kiloListesi)
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"g")
plt.xlabel("yas")
plt.ylabel("kilo")
plt.title("kilonun yaşa göre değişim grafiği")
```

```python
numpyDizisi1 = np.linspace(0,10,20)
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 3
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"r.-")
```

```python
plt.subplot(1,2,1)
plt.plot(numpyDizisi1,numpyDizisi2,"r.-")
plt.subplot(1,2,2)
plt.plot(numpyDizisi2,numpyDizisi1,"g--")
```

```python
myFigure = plt.figure()
figureAxes = myFigure.add_axes([0.5,0.5,0.6,0.5]) #boş ekrana tablo çizdirmeye yarar
figureAxes.plot(numpyDizisi1,numpyDizisi2,"r")
figureAxes.set_xlabel("x ekseni veri ismi")
figureAxes.set_ylabel("y ekseni veri ismi")
figureAxes.set_title("grafik ismi")
```

```python
numpyDizisiDeneme = np.linspace(0,10,20)
numpyDizisiDeneme2 = numpyDizisiDeneme **3 
plt.plot(numpyDizisiDeneme,numpyDizisiDeneme2,"g.-")
```

```python
#birden fazla grafik kullanılacagi zaman kullanilir
plt.subplot(1,2,1) #1 tane sıranın, 2 kolonunun 1. tablosunu belirtir 
plt.plot(numpyDizisiDeneme,numpyDizisiDeneme2,"r.-")
plt.subplot(1,2,2) # 1 tane sıranın, 2 kolonunun 2. tablosunu belirtir
plt.plot(numpyDizisi1,numpyDizisi2,"g--")
```

```python
#grafik içinde grafik yapma islemi
myFigure2 = plt.figure()
eksen1 = myFigure2.add_axes([0.1,0.1,0.7,0.7])
eksen2 = myFigure2.add_axes([0.2,0.4,0.3,0.3])

eksen1.plot(numpyDizisiDeneme,numpyDizisiDeneme2,"g")
eksen1.set_xlabel("X Ekseni")
eksen1.set_ylabel("Y Ekseni")
eksen1.set_title("ana grafik basligi")

eksen2.plot(numpyDizisiDeneme,numpyDizisiDeneme2,"b.-")
eksen2.set_xlabel("X Ekseni")
eksen2.set_ylabel("Y Ekseni")
eksen2.set_title("ana grafik basligi")


```

```python
plt.subplot(1,2,1)
plt.plot(numpyDizisiDeneme,numpyDizisiDeneme2)

plt.subplot(1,2,2)
```

```python
(myFigure, myFigure2) = plt.subplots(nrows=1, ncols=2)
type(myFigure2)

for myFigure in myFigure2:
    eksen1.plot(numpyDizisiDeneme,numpyDizisiDeneme2,"g")
    eksen1.set_xlabel("x ekseni")

plt.tight_layout() #tablolar arasini esitler
myFigure2
```

GÖRSEL GELİSTİRME VE KAYIT

```python
yeniFigure = plt.figure(figsize=(3,3)) #figsize komutu, tablo büyüklügünü ayarlar, dpi ayarı ise, tablo çözünürlügünü ayarlar
yeniEksen = yeniFigure.add_axes([0.1,0.1,0.9,0.9])
yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 2,"g.-",label = "numpyDizisi ** 2")
yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 3, label = "numpyDizisi ** 3")
yeniEksen.legend() #çizgilere eklenen label'ları gösterir
```

```python
yeniFigure.savefig("benimfigur.png",dpi = 200) #yapılan tabloyu kaydetmeye yarar. kaydedilecek tablo adı ve çözünürlük belirtilir.
```
